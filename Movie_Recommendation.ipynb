```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# difflib is used to search the movie name even if wrongly spelled name is entered, it will use that vry name and look for th closest match in the dataset.
```

```python
# Data collection and preprocessing
movies_data = pd.read_csv('/content/movies.csv')
movies_data.head()
```

```python
# creating content based recommendation system using genre, keywords, tagline, cast and director
selected_features = ['genres','keywords','tagline','cast','director']
# Replacing null values to null string
for feature in selected_features:
  movies_data[feature]= movies_data[feature].fillna('')
print(selected_features)
```

```python
# Combining all the features in one feature
combined_data = movies_data['genres']+ movies_data['keywords']+movies_data['tagline']+movies_data['director']+movies_data['cast']
```

```python
print(combined_data)
```

```python
# Now need to convert all the features in numerical form using tfdifvectorizer
vectorizer= TfidfVectorizer()
feature_vector = vectorizer.fit_transform(combined_data)
print(feature_vector)
```

```python
# Similarity score for all feature vectors
similarity = cosine_similarity(feature_vector)
print(similarity)
```

```python
similarity.shape
```

```python
# Here, for each movie out of 4803, it is comoared with every one and a score is obtained for each one, thus obtained a matrix of 4803X4803
```

```python
# Getting the movie name from user
user_input = input("Enter your favorite movie name:")
```

```python
# List of all movie names
list_of_movies = movies_data['title']
print(list_of_movies)
```

```python
# Finding closest match of uyser input in movie list
find_closest_match = difflib.get_close_matches(user_input, list_of_movies)
print(find_closest_match)
```

```python
# For the closest match
close_match = find_closest_match[0]
print("Did you mean:", close_match)
```

```python
# Finding the index of movie with title
index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
# getting a list of similar movies
similarity_score  =list(enumerate(similarity[index_of_movie]))
print(similarity_score)
len(similarity_score)
```

```python
# All movies with their similarity score is in the above list. But nit all the movies are desired. We wantonly ones with highest score.
```

```python
# Sorting movies nased on similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
# Here, sorted arrange the list in ascending order, so reverse is used to sort it in descending order. Key lamba uses to show that x =1 is to be sorted.
print(sorted_similar_movies)
```

```python
# printing names of top 30 movies 
print('Youll also like:')
i =1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index  =movies_data[movies_data.index==index]['title'].values[0]
  if (i <= 30):
    print(i,'.', title_from_index)
    i+=1
  
```
