```python
book_pivot = final_rating.pivot_table(columns = 'user_id', index= 'title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace =True)
```

```python
book_pivot
```

```python
# NN nearest neighbour algorithm for clustering will be used now. It will find the points with least distances and put them in a cluster.
```

```python
# There are lot of 0 in the book pivot which will make the calculation much time taking, so in order to avoid this,
# Sparse matrix is used.
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = 'brute')
# Generally, model just use some points and skip some, but when mentioned brute, it will calculate the distance for each point, and then will cluster.
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1))
# To get an entire row for a book, iloc was used,
# Reshape(1,-1) used to get a row of order1 * 888
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
  book_id= np.where(book_pivot.index==book_name)[0][0]
  distances, suggestions = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1))
  for i in range(len(suggestions)):
    print( "Suggestions for", book_name, "are", book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
